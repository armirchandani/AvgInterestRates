What the Loans mean:

Preapproval: A preliminary assessment of your borrowing capacity.

Origination: The full process of applying for and receiving a loan.

Purchased loan: A loan that has been sold by the originating lender to another party.

```python
#Labeling loans
df["loan_approval_status"] = df["action_taken_text"].map({
    "Loan originated": "Yes",
    "Purchased loan": "Yes",
    "Preapproval request approved but not accepted": "No - Application not approved",
    "Application approved but not accepted": "No - Application not approved"
})
df["loan_approval_status"].value_counts()
```

```python
df["average_interest_rate"].unique()
```

```python
# Check data set to see if they are all numeric
df["average_interest_rate"] = pd.to_numeric(df["average_interest_rate"], errors="coerce")

# Define interest rate thresholds and labels
interest_rates = [0, 3, 4, 5, 6, 7, 8, 9, 10]
rate_labels = [
    "Below 3%",
    "3.00%–3.99%",
    "4.00%–4.99%",
    "5.00%–5.99%",
    "6.00%–6.99%",
    "7.00%–7.99%",
    "8.00%–8.99%",
    "9.00% and above"
]

# Categorize rates into defined ranges
df["interest_rate_range"] = pd.cut(
    df["average_interest_rate"],
    bins=interest_rates,
    labels=rate_labels,
    right=False #boolean
)

# View the distribution
df["interest_rate_range"].value_counts(sort=False)
```

# Current average mortgage interest rates in the United States (as of July 19, 2025):

# 30-year fixed mortgage: **6.77%**
# 15-year fixed mortgage: **6.03%**

```python
# Filter DataFrame for "Black or African American"
df_black = df[df["derived_race_or_ethnicity"] == "Black or African American"]

# Count how many fall into each interest rate range for this group
rate_counts_black = df_black["interest_rate_range"].value_counts(sort=False)

# Plotting
plt.figure(figsize=(10, 6))
rate_counts_black.plot(kind='bar', color="#4C72B0", edgecolor='black')
plt.title("Distribution of Average Interest Rates\n(Black or African American)", fontsize=14)
plt.xlabel("Interest Rate Range", fontsize=12)
plt.ylabel("Number of Records", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Percentage of Housing Loans with Interest Rates over 6.77%

```python
# Ensure average_interest_rate is numeric
df["average_interest_rate"] = pd.to_numeric(df["average_interest_rate"], errors="coerce")

# Filter out rows with missing rates and for Black or African American group
df_black = df[(df["derived_race_or_ethnicity"] == "Black or African American") &
              (df["average_interest_rate"].notna())]

# Calculate counts above and below the benchmark for Black applicants
above_benchmark = (df_black["average_interest_rate"] > 6.77).sum()
below_benchmark = (df_black["average_interest_rate"] <= 6.77).sum()

# Data for pie chart
counts = [above_benchmark, below_benchmark]
labels = [f"> 6.77% ({above_benchmark})", f"≤ 6.77% ({below_benchmark})"]
colors = ["#FF6F61", "#4C72B0"]

# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(
    counts,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    wedgeprops={"edgecolor": "black"},
    textprops={"fontsize": 14}
)
plt.title("Percentage of Housing Loans with Interest Rate Above 6.77%\n(Black or African American)", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure average_interest_rate is numeric
df["average_interest_rate"] = pd.to_numeric(df["average_interest_rate"], errors="coerce")

# Filter for Black or African American and Female/Male applicants
df_filtered = df[
    (df["derived_race_or_ethnicity"] == "Black or African American") &
    (df["derived_sex"].isin(["Female", "Male"]))
].copy()

# Calculate average interest rate by sex
avg_interest = df_filtered.groupby("derived_sex")["average_interest_rate"].mean()

# Plot bar chart
plt.figure(figsize=(6,4))
ax = avg_interest.plot(kind='bar', color=["#FF69B4", "#1E90FF"])
plt.title("Average Interest Rate for Black or African American Applicants: Female vs Male")
plt.ylabel("Average Interest Rate (%)")
plt.xlabel("Sex")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add the average rate values on top of bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}%",
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

Average Interest Rates Based on Race

```python
df["derived_race_or_ethnicity"].unique()
```

```python
# Ensure interest rates are numeric
df["average_interest_rate"] = pd.to_numeric(df["average_interest_rate"], errors="coerce")

# Define races and sexes
races = ['Black or African American', 'Asian', 'Hispanic or Latino', 'White']
sexes = ['Female', 'Male']

# Filter data
df_filtered = df[
    (df["derived_race_or_ethnicity"].isin(races)) &
    (df["derived_sex"].isin(sexes))
].copy()

# Group and calculate mean interest rates
grouped = df_filtered.groupby(['derived_race_or_ethnicity', 'derived_sex'])['average_interest_rate'].mean().unstack()

# Plot settings
x = np.arange(len(races))  # label locations
width = 0.35  # bar width

fig, ax = plt.subplots(figsize=(10,6))

bars_female = ax.bar(x - width/2, grouped['Female'], width, label='Female', color='#FF69B4')
bars_male = ax.bar(x + width/2, grouped['Male'], width, label='Male', color='#1E90FF')

# Labels and title
ax.set_xlabel('Race/Ethnicity')
ax.set_ylabel('Average Interest Rate (%)')
ax.set_title('Average Interest Rate by Race and Sex')
ax.set_xticks(x)
ax.set_xticklabels(races, rotation=0)
ax.legend()

ax.grid(axis='y', linestyle='--', alpha=0.7)

# Attach data labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=11)

add_labels(bars_female)
add_labels(bars_male)

plt.tight_layout()
plt.show()
```
